## Binomial distribution

In probability theory and statistics, the binomial distribution is the discrete probability distribution that gives only two possible results in an experiment; for example, when you flip a coin, it can be either head or tails.

```python
from scipy.stats import binom

# Define all the probability functions related to this distribution

def dbinom(x, size, prob = 0.5):
    """
    Calculates the point estimate of the binomial distribution
    """
    result = binom.pmf(k = x, n = size, p = prob, loc = 0)

    return result

def pbinom(q, size, prob = 0.5):
    """
    Calculates the cumulative of the binomial distribution
    """
    result = binom.cdf(k = q, n = size, p = prob, loc = 0)

    return result

def qbinom(p, size, prob = 0.5):
    """
    Calculates the quantile function from the binomial distribution
    """
    result = binom.ppf(q = p, n = size, p = prob, loc = 0)

    return result

def rbinom(n, size, prob = 0.5):
    """
    Generates random variables from the binomial distribution
    """
    result = binom.rvs(n = size, p = prob, size = n)

    return result
```

```python
# Binomial(10, 0.2) distribution, where n = 10 is the number of trials and p = 0.2 is the probability of success in a single trial
import numpy as np

np.random.seed(42)

print(f"Probability that a Binomial(10, 0.2) takes the value 2: {dbinom(2, size = 10, prob = 0.2)}")
print(f"Probability that a Binomial(10, 0.2) will take a value less than 2: {pbinom(2, size = 10, prob = 0.2)}")
print(f"Which value of a Binomial(10, 0.2) has a cumulative probability of 0.9? {qbinom(0.9, size = 10, prob = 0.2)}")
print(f"Generate 2000 random values from a Binomial(10, 0.2) distribution: {rbinom(2000, size = 10, prob = 0.2)}")
```

```python
import matplotlib.pyplot as plt

np.random.seed(42)

size = 10
prob = 0.2
data = rbinom(2000, size = 10, prob = 0.2)

pmf = dbinom(data, size = size, prob = prob)
cdf = pbinom(data, size = size, prob = prob)
ppf = qbinom(data, size = size, prob = prob)

fig, axis = plt.subplots(3, 1, figsize = (5, 7))

axis[0].stem(data, pmf, "r-")
axis[1].stem(data, cdf, "b-")
axis[2].stem(data, ppf, "g-")

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import uniform

# Define all the probability functions related to the uniform distribution

def dunif(x, min_val=0, max_val=10):
    """
    Calculates the point estimate (probability density function) of the uniform distribution
    """
    result = uniform.pdf(x, loc=min_val, scale=max_val-min_val)
    return result

def punif(q, min_val=0, max_val=10):
    """
    Calculates the cumulative distribution function of the uniform distribution
    """
    result = uniform.cdf(q, loc=min_val, scale=max_val-min_val)
    return result

def qunif(p, min_val=0, max_val=10):
    """
    Calculates the quantile function (inverse CDF) of the uniform distribution
    """
    result = uniform.ppf(p, loc=min_val, scale=max_val-min_val)
    return result

def runif(n, min_val=0, max_val=10):
    """
    Generates random variables from the uniform distribution
    """
    result = uniform.rvs(loc=min_val, scale=max_val-min_val, size=n)
    return result

# Uniform distribution [0, 1], where min_val = 0 and max_val = 1
import numpy as np

np.random.seed(42)

print(f"Probability density function values for the points [0, 0.5, 1]: {dunif([0, 0.5, 1], min_val=0, max_val=10)}")
print(f"Cumulative probability that a Uniform(0, 1) takes a value less than 0.5: {punif(0.5, min_val=0, max_val=10)}")
print(f"The value of a Uniform(0, 1) distribution that has a cumulative probability of 0.9: {qunif(0.9, min_val=0, max_val=10)}")
print(f"Generate 2000 random values from a Uniform(0, 1) distribution: {runif(2000, min_val=0, max_val=10)}")

import matplotlib.pyplot as plt

# Generate random data from a uniform distribution
np.random.seed(42)
min_val = -1
max_val = 2
data = runif(2000, min_val=min_val, max_val=max_val)

# Calculate the quantile function (PPF) for a range of cumulative probabilities
quantiles = np.linspace(0.01, 0.99, 1000)  # Increase number of points for smoother plot
ppf_values = qunif(quantiles, min_val=min_val, max_val=max_val)

# Create subplots
fig, axis = plt.subplots(3, 1, figsize=(10, 12))

# Plot the probability density function (PDF)
x_values = np.linspace(min_val, max_val, 1000)  # Adjust x values to cover wider range
axis[0].plot(x_values, dunif(x_values, min_val=min_val, max_val=max_val), "r-", linewidth=2)
axis[0].fill_between(x_values, dunif(x_values, min_val=min_val, max_val=max_val), color="red", alpha=0.2)
axis[0].set_title('Probability Density Function (PDF)', fontsize=16)

# Plot the cumulative distribution function (CDF)
axis[1].plot(x_values, punif(x_values, min_val=min_val, max_val=max_val), "b-", linewidth=2)
axis[1].fill_between(x_values, punif(x_values, min_val=min_val, max_val=max_val), color="blue", alpha=0.2)
axis[1].set_title('Cumulative Distribution Function (CDF)', fontsize=16)

# Plot the quantile function (PPF)
axis[2].plot(quantiles, ppf_values, "g-", linewidth=2)
axis[2].fill_between(quantiles, quantiles, ppf_values, color="green", alpha=0.2)
axis[2].set_title('Quantile Function (PPF)', fontsize=16)

# Set common labels
for ax in axis:
    ax.set_xlabel('Value', fontsize=14)
    ax.set_ylabel('Probability', fontsize=14)
    ax.grid(True)
    ax.set_xticks(np.arange(min_val, max_val+1, 1))  # Add more ticks for better visualization

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```
